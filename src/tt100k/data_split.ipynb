```python
import os
import shutil
from tqdm import tqdm

# ============================
# 1) Cáº¥u hÃ¬nh Ä‘Æ°á»ng dáº«n
# ============================
IMG_DIR = "./Dataset/Detect/data_detect_tt100k/images/test"     # thÆ° má»¥c áº£nh nguá»“n
LBL_DIR = "./Dataset/Detect/data_detect_tt100k/labels/test"     # thÆ° má»¥c label nguá»“n

OUT_IMG_DIR = "/home/pi5/TrafficSign/TT100K/Dataset/Detect/split/images/test"
OUT_LBL_DIR = "/home/pi5/TrafficSign/TT100K/Dataset/Detect/split/labels/test"

N = 1000   # sá»‘ lÆ°á»£ng áº£nh muá»‘n copy

# ============================
# 2) Táº¡o thÆ° má»¥c output
# ============================
os.makedirs(OUT_IMG_DIR, exist_ok=True)
os.makedirs(OUT_LBL_DIR, exist_ok=True)

# ============================
# 3) Láº¥y danh sÃ¡ch áº£nh
# ============================
img_files = [f for f in os.listdir(IMG_DIR) if f.endswith(".jpg") or f.endswith(".png")]
img_files = sorted(img_files)[:N]  # láº¥y 1000 áº£nh Ä‘áº§u tiÃªn

print(f"ðŸ“¸ Tá»•ng sá»‘ áº£nh láº¥y: {len(img_files)}")

# ============================
# 4) Copy áº£nh + label tÆ°Æ¡ng á»©ng
# ============================
for img_name in tqdm(img_files, desc="Copying"):
    img_src = os.path.join(IMG_DIR, img_name)
    img_dst = os.path.join(OUT_IMG_DIR, img_name)

    # copy áº£nh
    shutil.copy2(img_src, img_dst)

    # tÃ¬m label tÆ°Æ¡ng á»©ng
    stem = os.path.splitext(img_name)[0]
    lbl_name = stem + ".txt"
    lbl_src = os.path.join(LBL_DIR, lbl_name)
    lbl_dst = os.path.join(OUT_LBL_DIR, lbl_name)

    # copy label (náº¿u cÃ³)
    if os.path.exists(lbl_src):
        shutil.copy2(lbl_src, lbl_dst)
    else:
        # khÃ´ng cÃ³ label â†’ táº¡o file rá»—ng Ä‘á»ƒ giá»¯ format
        with open(lbl_dst, "w") as f:
            pass
```
